Section 4 — Vectorization vs Loops

Task 4.1: Loop vs Vectorized

```python
# TODO:
# Create large array X of size 1,000,000

X = np.random.randn(1_000_000)
```

```python
# TODO:
# Normalize using Python loop

# HINT:
# - Time it
import time

X_loop = X.copy()

start = time.time()

total = 0
for x in X_loop:
    total += x

for i in range(len(X_loop)):
    X_loop[i] = X_loop[i] / total

end = time.time()

print("Loop time:", end - start)


```

```python
# TODO:
# Normalize using vectorization

X_vec = X.copy()

start = time.time()

X_vec = X_vec / X_vec.sum()

end = time.time()

print("Vectorized time:", end - start)
```

Why is vectorization faster?


Task 4.2: Pairwise Distance (FAANG Classic)

```python
# TODO:
# Compute pairwise Euclidean distance matrix without loops

# HINT:
# - Use (x - y)^2 expansion
# - Broadcasting is key
import numpy as np
def pairwise_distance(X):
    X = np.asarray(X, dtype=float)
    if X.ndim == 1:
        X = X.reshape(-1, 1)
    sq = np.sum(X * X, axis=1)
    dist2 = sq[:, None] + sq[None, :] - 2 * (X @ X.T)
    dist2 = np.maximum(dist2, 0.0)
    return np.sqrt(dist2)

X = np.random.randn(5, 3)

print(pairwise_distance(X))

print(pairwise_distance(X))
```

Task 5.1: Softmax

```python
# TODO:
# Implement naive softmax
def softmax_naive(X):
    ex = np.exp(X)
    return ex / ex.sum(axis=1, keepdims=True)

def softmax_stable(X):
    Z = X - X.max(axis=1, keepdims=True)
    ex = np.exp(Z)
    return ex / ex.sum(axis=1, keepdims=True)

Xbig = np.array([[1000.0, 1001.0, 1002.0]])
try:
    print('naive:', softmax_naive(Xbig))
except FloatingPointError as e:
    print('overflow:', e)

print('stable:', softmax_stable(Xbig))
print('rowsum:', softmax_stable(Xbig).sum(axis=1))
```

Section 7 — Performance & Memory

Task 7.1: In-Place Operations

```python
# TODO:
# Compare in-place vs out-of-place operations
import numpy as np
import time

def compare_inplace_vs_outofplace(n=2_000_000):
    rng = np.random.default_rng(7)
    X = rng.standard_normal(n)
    Y = X.copy()

    start = time.perf_counter()
    Z = X + 1.0
    out_time = time.perf_counter() - start

    start = time.perf_counter()
    Y += 1.0
    in_time = time.perf_counter() - start

    print(f"Out-of-place time: {out_time:.6f} seconds")
    print(f"In-place time    : {in_time:.6f} seconds")

    return out_time, in_time

compare_inplace_vs_outofplace()


```

Section 8 — Mini Case Study

```python
# TODO:
# Given X (10000, 100):
# - Normalize features
# - Compute covariance
# - Extract top-k eigenvectors

import numpy as np

def pca_project(X, k=5, ddof=1, eps=1e-12):
    X = np.asarray(X, dtype=float)
    n = X.shape[0]

    mu = X.mean(axis=0, keepdims=True)
    sigma = X.std(axis=0, ddof=0, keepdims=True)
    sigma = np.maximum(sigma, eps)
    Xn = (X - mu) / sigma

    C = (Xn.T @ Xn) / (n - ddof)

    eigvals, eigvecs = np.linalg.eigh(C)
    idx = np.argsort(eigvals)[-k:][::-1]

    W = eigvecs[:, idx]
    Z = Xn @ W

    return Z, W, eigvals[idx], mu, sigma, C

rng = np.random.default_rng(8)
X = rng.standard_normal((10_000, 100))

Z, W, top_eigvals, mu, sigma, C = pca_project(X, k=5)

print("C shape:", C.shape)
print("W shape:", W.shape)
print("Z shape:", Z.shape)
print("top eigvals:", top_eigvals)
```
